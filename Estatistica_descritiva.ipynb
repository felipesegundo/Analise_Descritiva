#Projeto - Estatística descritiva
##Este projeto tem por objetivo analisar as medidas de frequência, medidas de tendência central e medidas de dispersão de uma base de dados referente a prova do Enem realizado no Estado de São Paulo em 2019.

##Importação das bibliotecas essênciais

```r
library(dplyr)
if(!require(sampling)) install.packages("sampling")
if(!require(TeachingSampling)) install.packages("TeachingSampling")
library(sampling)
library(TeachingSampling)
df = read.csv("/content/enem2019_tratado.csv", sep = ",")
```

##Filtrando os dados para obter a cidade de 'Sorocaba'

```r
options(repr.plot.width = 22, repr.plot.height = 8)
```

```r
enem_sorocaba <- df %>% filter(NO_MUNICIPIO_RESIDENCIA=="Sorocaba")
```

##Tabela de frequência absoluta por idade de candidatos na cidade de 'Sorocaba'

```r

freq_abs = (table(enem_sorocaba$NU_IDADE))
freq_abs
```

##Tabela de frquencia absoluta por idade em porcentagem

```r
freq_rel = prop.table(freq_abs)
freq_rel
```

```r
#Tabela de frquencia absoluta por idade em porcentagem
p_freq_rel = prop.table(freq_abs)*100
p_freq_rel
```

## Juntando a frequência relativa e a frequência percentual com suas respectivas somas.


```r
# Juntando a frequência relativa e a frequência percentual com suas respectivas somas.
freq_rel <- c(freq_rel, sum(freq_rel))
p_freq_rel <- c(p_freq_rel, sum(p_freq_rel))

# Tabela final com todos os valores
tabela_final <- cbind(freq_abs, freq_rel = round(freq_rel, digits = 5),
p_freq_rel = round(p_freq_rel, digits = 2))
```

```r
(tabela_final)
```

##Construindo classes de frequencias

```r
#construindo classes de frquencias
intervalo_de_classes = seq(10, 75, 5)
intervalo_de_classes
```

```r
#construindo as classes
tabela_classes <- table(cut(enem_sorocaba$NU_IDADE, breaks=intervalo_de_classes, right=FALSE))
tabela_classes
```

##Gráficos de frequencia

```r
#histograma
par(mar = c(5, 5, 4, 2) + 0.1) # margem(c(cima, direita, baixo e esquerda)
histograma = hist(enem_sorocaba$NU_IDADE,
                cex.main = 3.5, #tamanho do titulo
                col = "skyblue", #cor das barras
                main = "Frequência das idades", #titulo
                cex.axis = 1.5, #tamanho dos valores de x e y
                cex.lab = 2, #tamanhos dos rotulos de x e y
                xlab = "Idades", #rotulos de x
                ylab = "Frenquencia das idades", #rotulos de y
                border = "white") #cor das linhas que separa as barras

text(x = histograma$mids, y = histograma$counts, labels = histograma$counts, pos = 3, cex = 1, col = "black") #adiciona valores para as barras
```

```r
#Gráfico de ogiva
#frequencia acumulada
frequencia_relativa_classes = prop.table(table(cut(enem_sorocaba$NU_IDADE, breaks = c(intervalo_de_classes))))
frequencia_relativa_classes
```

```r
frequencia_acumulada = cumsum(tabela_classes)[seq_along(intervalo_de_classes)]
frequencia_acumulada
```

```r
par(mar = c(5, 5, 4, 2) + 0.1)
plot(intervalo_de_classes, frequencia_acumulada, type = "o",
  cex.main = 3.5, #tamanho do titulo
                col = "red", #cor das barras
                main = "Frequência acumulada das idades", #titulo
                cex.axis = 1.5, #tamanho dos valores de x e y
                cex.lab = 2, #tamanhos dos rotulos de x e y
                xlab = "Idades", #rotulos de x
                ylab = "Soma das frequencias", #rotulos de y
                border = "white") #cor das linhas que separa as barras
```

##Analisando a distribuição das notas de matemática

```r
# Configurando o ambiente gráfico para ter 2 gráficos lado a lado
par(mfrow = c(1, 2)) # 1 linha, 2 colunas

# Configurando as margens dos gráficos
par(mar = c(5, 5, 4, 2) + 0.1)

# Primeiro gráfico: histograma com linha de densidade
hist(enem_sorocaba$NOTA_MT, probability = TRUE,
     cex.main = 3.5, # tamanho do título
     col = "skyblue", # cor das barras
     main = "Distribuição das notas de matemática", # título
     cex.axis = 1.5, # tamanho dos valores de x e y
     cex.lab = 2, # tamanhos dos rótulos de x e y
     xlab = "Notas", # rótulos de x
     ylab = "Porcentagem", # rótulos de y
     border = "white") # cor das linhas que separam as barras
lines(density(df$NOTA_MT), col = "red")

# Segundo gráfico: gráfico Q-Q
qqnorm(df$NOTA_MT)
qqline(df$NOTA_MT, col = "red")

# Restaurando a configuração padrão do layout
par(mfrow = c(1, 1))
```

##Distribuição da nota de redação

```r
par(mfrow = c(1, 2)) # 1 linha, 2 colunas

# Configurando as margens dos gráficos
par(mar = c(5, 5, 4, 2) + 0.1)

# Primeiro gráfico: histograma com linha de densidade
hist(enem_sorocaba$NOTA_REDACAO, probability = TRUE,
     cex.main = 3.5, # tamanho do título
     col = "skyblue", # cor das barras
     main = "Distribuição das notas de matemática", # título
     cex.axis = 1.5, # tamanho dos valores de x e y
     cex.lab = 2, # tamanhos dos rótulos de x e y
     xlab = "Notas", # rótulos de x
     ylab = "Porcentagem", # rótulos de y
     border = "white") # cor das linhas que separam as barras
lines(density(df$NOTA_REDACAO), col = "red")

# Segundo gráfico: gráfico Q-Q
qqnorm(df$NOTA_REDACAO)
qqline(df$NOTA_REDACAO, col = "red")

# Restaurando a configuração padrão do layout
par(mfrow = c(1, 1))
```
